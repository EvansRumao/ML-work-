Merging Joining and concatenation

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries) # By default, it performs an inner join, which means it will only include rows with matching employee_id in both DataFrames.
```

```python
pd.merge(employees,salaries,on='employee_id',how ='inner') # Inner join: Only include rows with matching employee_id in both DataFrames.
```

```python
pd.merge(employees,salaries,on='employee_id',how ='outer') # Outer join: Include all rows from both DataFrames, filling in NaN for missing matches.
```

```python
pd.merge(employees,salaries,on='employee_id',how ='left') # Left join: Include all rows from the left DataFrame (employees) and matching rows from the right DataFrame (salaries). Fill in NaN for missing matches in the right DataFrame.
```

```python
pd.merge(employees,salaries,on='employee_id',how ='right') # Right join: Include all rows from the right DataFrame (salaries) and matching rows from the left DataFrame (employees). Fill in NaN for missing matches in the left DataFrame.
```

Concatenation of 2 DataFrames

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2]) # Concatenate df1 and df2 vertically (stacking rows). By default, it ignores the index and creates a new one.
```

```python
pd.concat([df2, df1])# Concatenate df2 and df1 vertically (stacking rows). By default, it ignores the index and creates a new one. The order of DataFrames in the list determines the order of rows in the resulting DataFrame.
```

```python
pd.concat([df1,df2],axis=1) # Concatenate df1 and df2 horizontally (stacking columns). By default, it ignores the index and creates a new one. The order of DataFrames in the list determines the order of columns in the resulting DataFrame.
```

Joining 2 data frames

```python
df1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

# Second DataFrame
df2 = pd.DataFrame({
    'score': [85, 90, 75]
}, index=[2, 3, 4])
```

```python
df1
```

```python
df2
```

```python
df1.join([df2],how="inner") # Join df1 and df2 based on their indices. By default, it performs a left join, which means it will include all rows from df1 and matching rows from df2. Fill in NaN for missing matches in df2.
```

```python
df1.join([df2],how="outer")
```

```python
df1.join([df2],how="left")
```

```python
df2.join([df1],how="inner")
```
