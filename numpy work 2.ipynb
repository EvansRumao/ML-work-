Deep and shallow copy

```python
a=np.arange(1,10)
a
```

```python
slice=a[:5]
slice
```

```python
slice=slice+10
print(slice)
print(a)
```

```python
b=a
b[0]=100
print(b)
print(a)
```

```python
a=np.arange(1,10)
b=a.copy()
b[0]=100
print(b)
print(a)
```
